```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [15, 30, 45, 10]
random_data = np.random.normal(5, 2, 100)
```

```python
# Bar Plot
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color='orange')
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='green')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8, 5))
plt.hist(random_data, bins=10, color='purple', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(8, 5))
plt.boxplot(random_data)
plt.title('Box Plot')
plt.ylabel('Value')
plt.show()
```
