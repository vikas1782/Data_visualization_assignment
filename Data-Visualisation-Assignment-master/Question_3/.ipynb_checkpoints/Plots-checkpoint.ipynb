```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('iris')
print(data.head())
```

```python
# Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='petal_length', y='petal_width', hue='species')
plt.title("Scatter Plot of Petal Length vs. Petal Width")
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='species', y='sepal_length')
plt.title("Box Plot of Sepal Length Across Species")
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=data, x='species', y='petal_length')
plt.title("Violin Plot of Petal Length Across Species")
plt.show()
```

```python
# Pie Chart
species_counts = data['species'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Samples per Species")
plt.show()
```

```python
# Stacked Bar Plot
average_lengths = data.groupby('species')[['sepal_length', 'petal_length']].mean()
average_lengths.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title("Average Sepal and Petal Lengths per Species")
plt.ylabel("Average Length (cm)")
plt.show()
```

```python
# Area Plot (Cumulative Sepal Length Distribution)
data_sorted = data.sort_values('sepal_length')
plt.figure(figsize=(10, 6))
plt.fill_between(range(len(data_sorted)), data_sorted['sepal_length'].cumsum(), color='skyblue', alpha=0.5)
plt.title("Cumulative Distribution of Sepal Lengths")
plt.xlabel("Sample Index")
plt.ylabel("Cumulative Sepal Length")
plt.show()
```
