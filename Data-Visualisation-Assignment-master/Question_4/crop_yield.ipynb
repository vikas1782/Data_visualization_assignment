```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('crop_yield.csv')
print(data.head())
```

```python
maize_data = data[(data['Crop'] == 'Maize') & (data['State'] == 'Karnataka')]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(maize_data['Crop_Year'], maize_data['Production'], marker='o', color='b', label='Maize Production')
plt.title("Maize Production Over the Years in Karnataka")
plt.xlabel("Year")
plt.ylabel("Production (in Tons)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Scatter Plot ----> Fertilizer vs Crop yield

plt.figure(figsize=(10, 6))
plt.scatter(data['Fertilizer'], data['Yield'], color='g', alpha=0.6)
plt.title("Relationship Between Fertilizer Usage and Crop Yield")
plt.xlabel("Fertilizer (in kg/ha)")
plt.ylabel("Yield (in kg/ha)")
plt.text(50, 1000, "Higher Fertilizer Usage -> Higher Yield", fontsize=12, color='red')
plt.grid(True)
plt.show()
```

```python
# Barplot ---> Average Production by crop

avg_production_by_crop = data.groupby('Crop')['Production'].mean().sort_values()

plt.figure(figsize=(12, 8))
avg_production_by_crop.plot(kind='bar', color='orange')
plt.title("Average Production by Crop Type")
plt.xlabel("Crop Type")
plt.ylabel("Average Production (in Tons)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram ---> Rainfall Distribution

plt.figure(figsize=(10, 6))
plt.hist(data['Annual_Rainfall'], bins=15, color='skyblue', edgecolor='black')
plt.title("Annual Rainfall Distribution")
plt.xlabel("Annual Rainfall (in mm)")
plt.ylabel("Frequency")
plt.text(1000, 50, "Most regions receive 500-1000 mm rainfall", fontsize=12, color='blue')
plt.legend(["Rainfall Distribution"])
plt.show()
```

```python
# Pie Chart ----> Percentage of area used by crops.
area_by_crop = data.groupby('Crop')['Area'].sum()
plt.figure(figsize=(8, 8))
plt.pie(area_by_crop, labels=area_by_crop.index, autopct='%1.1f%%', startangle=140)
plt.title("Area Used by Different Crops")
plt.show()
```

```python
# Multi Line Plot

crops_data = data[(data['Crop'] == 'Maize') | (data['Crop'] == 'Rice')]

plt.figure(figsize=(10, 6))
for crop in crops_data['Crop'].unique():
    crop_data = crops_data[crops_data['Crop'] == crop]
    plt.plot(crop_data['Crop_Year'], crop_data['Production'], marker='o', label=f"{crop} Production")

plt.title("Production Trends for Maize and Rice")
plt.xlabel("Year")
plt.ylabel("Production (in Tons)")
plt.legend()
plt.show()
```
